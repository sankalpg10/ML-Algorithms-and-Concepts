```python
#splitting data into training and testing sets

def train_test_split(dataframe, n):  #    0<n< 1
  k = len(dataframe)

  df_train = dataframe.iloc[:int(n*k),:]
  df_test = dataframe.iloc[int(n*k):,:]

  return df_train, df_test
```

```python
#function to get euclidean distance:

from math import sqrt

def euclidean_dist(row1,row2):

  k = len(row1)  #because we dont want to include the price column
  total_diff = 0
  for i in range(k):
    total_diff = total_diff + ((row1[i] - row2[i])**2)

    euclidean_dist = sqrt(total_diff)

    return euclidean_dist

```
